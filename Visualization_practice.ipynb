```python
import pandas as pd
import json
```

```python
# Opening JSON file
f = open('matches\\t20s_json\\211028.json')

# returns JSON object as a dictionary
data = json.load(f)
```

```python
# Extract official data
officials=[]

for off in data['info']['officials']:
    #print(,data['info']['officials'][off])
    for person in data['info']['officials'][off]:        
        officials_table={
    "match_id":1,
    "person_id":data['info']['registry']['people'][person],
    "official_type":off,
}
        officials.append(officials_table)

df_officials=pd.DataFrame(officials)

print(df_officials)
    
```

```python
#registry table data

df_registry=pd.DataFrame({
'person_id ':[],
'person_name':[]
})

for i in data['info']['registry']['people']:
    if data['info']['registry']['people'][i] not in df_registry['person_id ']:
        #print(i, data['info']['registry']['people'][i])
        new_row=pd.DataFrame({
            'person_id ':[data['info']['registry']['people'][i]],
            'person_name':[i]
        })
        df_registry = pd.concat([df_registry, new_row], ignore_index=True)

print(df_registry)
```

```python
#players

df_players=pd.DataFrame({
    "match_id":1 ,
    "person_id":[],
    "team_name":[]
})

for player in data['info']['players']:
    #print(player,data['info']['players'][player])
    for each_player in data['info']['players'][player]:
        df_players.loc[len(df_players)] = [1,data['info']['registry']['people'][each_player],player]

print(df_players)
```

```python
#for match_info in data['info']:
d=data['info']
match={
"match_id":1,
"city":d['city'],
"gender":d['gender'] , 
"match_type":d["match_type"] , 
"match_type_number" :d["match_type_number"], 
"overs":d["overs"] , 
"season":d["season"] , 
"team_type":d["team_type"] , 
"venue":d["venue"] ,
"team1":d["teams"][0],
"team2":d["teams"][1],
"toss_winner":d["toss"]["winner"] , 
"toss_decision":d["toss"]["decision"] , 
"winner":d["outcome"]["winner"] or None, 
"outcome_type":list(d["outcome"]["by"].keys())[0] if d["outcome"]["by"] else d["outcome"]["result"] , # 'runs', 'wickets', 'tie', 'no result', etc.
"outcome_value": d["outcome"].get("by") or None, # The number of runs or wickets in the outcome (NULL if not applicable)
"player_of_match":f"{', '.join([d["registry"]['people'].get(i) for i in d['player_of_match']])}" if len(d["player_of_match"]) > 1 else d["registry"]['people'].get("".join(d["player_of_match"])) , 
"balls_per_over":d["balls_per_over"]
    }

#print(match)

#match_details
df_match_details=pd.DataFrame(match)
print(df_match_details)
```

```python
# Extract deliveries data
deliveries = []
inning_count=0
for inning in data['innings']:
    inning_count+=1
    for over in inning['overs']:
        ball_count=0
        for delivery in over['deliveries']:
            ball_count+=1
            delivery_data = {
                'innings':inning_count,
                'team': inning['team'],
                'over': over['over'],
                'balls': ball_count,
                'batter': delivery['batter'],
                'bowler': delivery['bowler'],
                'non_striker': delivery['non_striker'],
                'runs_batter': delivery['runs']['batter'],
                'runs_extras': delivery['runs']['extras'],
                'runs_total': delivery['runs']['total'],
                'powerplayed': 'no',
                'powerplayed_type': 'none'
            }
            
            deliveries.append(delivery_data)

# Create DataFrame
df_deliveries = pd.DataFrame(deliveries)

print(df_deliveries)
```

```python
# Opening JSON file
f1 = open('matches\\tests_json\\225265.json')

# returns JSON object as a dictionary
data1 = json.load(f1)
```

```python
def get_registryID(personName,data):
    # Get the registry ID from the data
    return data['info']['registry']['people'][personName]
```

```python
deliveries, inning_count = [], 0

for inning in data1['innings']:
    inning_count += 1
    team, powerplays = inning['team'], {pp['type']: (int(pp['from']), int(pp['to'])) for pp in inning.get('powerplays', [])}
    miscounted = {int(k): v['balls'] for k, v in inning.get('miscounted_overs', {}).items()}

    for over in inning['overs']:
        over_num, balls_limit = over['over'], miscounted.get(over['over'], 6)
        
        for ball_count, delivery in enumerate(over['deliveries'], start=1):
            if ball_count > balls_limit: continue  # Skip extra miscounted deliveries
            powerplay_type = next((t for t, (s, e) in powerplays.items() if s <= over_num <= e), 'none')

            if len(delivery.get('wickets', {})) > 0:
                 player_out = get_registryID(delivery['wickets'][0].get('player_out'),data1) if delivery['wickets'][0].get('player_out') else None
                 kind = delivery['wickets'][0].get('kind', None)
                 fielders = ', '.join(get_registryID(f['name'],data1) for f in delivery['wickets'][0].get('fielders', [])) if delivery['wickets'][0].get('fielders') else None
            else:
                player_out, kind, fielders = None, None, None

            deliveries.append({
                'innings': inning_count, 
                'team': team, 
                'over': over_num, 
                'balls': ball_count,
                'batter': get_registryID(delivery['batter'],data1), 
                'bowler': get_registryID(delivery['bowler'],data1), 
                'non_striker': get_registryID(delivery['non_striker'],data1),
                'runs_batter': delivery['runs']['batter'], 
                'runs_extras': delivery['runs']['extras'],
                'runs_total': delivery['runs']['total'], 
                'powerplayed': 'yes' if powerplay_type != 'none' else 'no',
                'powerplayed_type': powerplay_type,
                'player_out': player_out, 
                'dismissal_kind': kind, 
                'fielders_involved': fielders 
            })

df_deliveries = pd.DataFrame(deliveries)
print(df_deliveries.loc[1:15])
```
