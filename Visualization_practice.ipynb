```python
import pandas as pd
import json
```

```python
# Opening JSON file
f = open('matches\\t20s_json\\211028.json')

# returns JSON object as a dictionary
data = json.load(f)
```

```python
data
```

```python
# Extract deliveries data
deliveries = []
for inning in data['innings']:
    for over in inning['overs']:
        for delivery in over['deliveries']:
            delivery_data = {
                'team': inning['team'],
                'over': over['over'],
                'batter': delivery['batter'],
                'bowler': delivery['bowler'],
                'non_striker': delivery['non_striker'],
                'runs_batter': delivery['runs']['batter'],
                'runs_extras': delivery['runs']['extras'],
                'runs_total': delivery['runs']['total']
            }
            deliveries.append(delivery_data)

# Create DataFrame
df_deliveries = pd.DataFrame(deliveries)

print(df_deliveries)
```

```python
# Extract official data
officials=[]

for off in data['info']['officials']:
    #print(,data['info']['officials'][off])
    for person in data['info']['officials'][off]:        
        officials_table={
    "match_id":1,
    "person_id":data['info']['registry']['people'][person],
    "official_type":off,
}
        officials.append(officials_table)

df_officials=pd.DataFrame(officials)

print(df_officials)
    
```

```python
#registry table data

df_registry=pd.DataFrame({
'person_id ':[],
'person_name':[]
})

for i in data['info']['registry']['people']:
    if data['info']['registry']['people'][i] not in df_registry['person_id ']:
        #print(i, data['info']['registry']['people'][i])
        new_row=pd.DataFrame({
            'person_id ':[data['info']['registry']['people'][i]],
            'person_name':[i]
        })
        df_registry = pd.concat([df_registry, new_row], ignore_index=True)

print(df_registry)
```

```python
#players

df_players=pd.DataFrame({
    "match_id":1 ,
    "person_id":[],
    "team_name":[]
})

for player in data['info']['players']:
    #print(player,data['info']['players'][player])
    for each_player in data['info']['players'][player]:
        df_players.loc[len(df_players)] = [1,data['info']['registry']['people'][each_player],player]

print(df_players)
```
